```python
import os
import random

def thanos_folder(folder_path, num_files):
    # 1- التأكد من وجود المجلد أو إنشاؤه
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder created at: {folder_path}")
    else:
        print(f"Folder exists at: {folder_path}")

    # 2- إنشاء عدد من الملفات
    for i in range(1, num_files + 1):
        file_path = os.path.join(folder_path, f"file_{i}.txt")
        with open(file_path, "w") as f:
            f.write(f"This is file {i}\n")
    print(f"{num_files} files created.")

    # 3- التحقق من عدد الملفات
    all_files = os.listdir(folder_path)
    print(f"Number of files in folder: {len(all_files)}")

    # 4- حذف نصف الملفات عشوائيًا
    files_to_delete = random.sample(all_files, len(all_files)//2)
    for file in files_to_delete:
        os.remove(os.path.join(folder_path, file))
    print(f"Deleted {len(files_to_delete)} files randomly.")

    # 5- التحقق من النتيجة النهائية
    remaining_files = os.listdir(folder_path)
    print(f"Remaining files: {len(remaining_files)}")
    print("Remaining files list:", remaining_files)

# مثال على الاستخدام
thanos_folder("test_folder", 10)
```
