```python
class Task1:
 
    def __init__(self, number):
        self.number = number

    def check_even_odd(self):
        if self.number % 2 == 0:
            return "زوجي"
        else:
            return "فردي"
num = 10
task1_obj = Task1(num)
result = task1_obj.check_even_odd()
print(f"الرقم {num} هو {result}.")
```
