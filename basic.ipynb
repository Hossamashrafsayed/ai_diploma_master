```python
def process_email(email):
    if email.count("@") != 1:
        return "Invalid email"
    username, rest = email.split("@")
    if "." not in rest:
        return "Invalid email"
    domain = rest.rsplit(".", 1)[0]
    if email.endswith(".com"):
        domain_type = "Commercial Domain"
    elif email.endswith(".edu"):
        domain_type = "Educational Domain"
    else:
        domain_type = "Other Domain"

    return {
        "Username": username,
        "Domain": domain,
        "Domain Type": domain_type
    }
email = "Amit_ml@gmail.edu"
print(process_email(email))
```

```python
encoded = "###!!@mocleW EPGTQ!!!6789"


core = "".join([c for c in encoded if c.isalpha() or c == " "]).strip()

first, second = core.split()
first_decoded = first[::-1]
decoded_message = f"{first_decoded} {second}"
print(decoded_message)
```

```python
encoded = "&&&**$gnirtS PLIO!!@1234"
core = "".join([c for c in encoded if c.isalpha() or c == " "]).strip()

first, second = core.split()
first_decoded = first[::-1]
vowel_map = {
    "I": "E",
    "O": "U"
}

second_decoded = "".join(vowel_map.get(c, c) for c in second)

decoded_message = f"{first_decoded} {second_decoded}"

print(decoded_message)
```
